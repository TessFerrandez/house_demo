```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/interim/houses.csv')
df.count()
```

```python
df = df[df.area.isnull() == False]
```

```python
df_rooms = df.copy()
df_rooms = df_rooms[df_rooms.rooms.isnull() == False]
df_rooms['area_per_room'] = df_rooms.area / df_rooms.rooms
np.round(df_rooms.area_per_room.mean(), 0)
```

```python
df.rooms.fillna(df.area / 25.0, inplace=True)
```

```python
df.land_area.fillna(0, inplace=True)
```

```python
df.count()
```

```python
df['is_condo'] = df.monthly_fee > 0
df[df.is_condo == False].head()
```

```python
df[df.is_condo].head()
```

```python
df['total_area'] = df.area + df.sup_area
df['price_per_sqm'] = df.price / df.area
df['price_per_tsqm'] = df.price / df.total_area
df['list_price'] = np.round((df.price * 100 / (100 + df.price_change)) / 1000, 0) * 1000
```

```python
df['region'].fillna("UNKNOWN", inplace=True)

def clean_region(region):
    u_region = region.upper()
    if "VIBY" in u_region:
        return "VIBY"
    if "NORRVIKEN" in u_region:
        return "NORRVIKEN"
    if "TÖJNAN" in u_region:
        return "TÖJNAN"
    if "EDSVIKEN" in u_region:
        return "EDSVIKEN"
    if "HELENELUND" in u_region:
        return "HELENELUND"
    if "ROTEBRO" in u_region:
        return "ROTEBRO"
    if "HÄGGVIK" in u_region:
        return "HÄGGVIK"
    if "FÅGELSÅNGEN" in u_region:
        return "FÅGELSÅNGEN"
    if "SILVERDAL" in u_region:
        return "SILVERDAL"
    if "TEGELHAGEN" in u_region:
        return "TEGELHAGEN"
    if "TÖRNSKOGEN" in u_region:
        return "TÖRNSKOGEN"
    if "SJÖBERG" in u_region:
        return "SJÖBERG"
    if "ROTSUNDA" in u_region:
        return "ROTSUNDA"
    if "KÄRRDAL" in u_region:
        return "KÄRRDAL"
    if "TUREBERG" in u_region:
        return "TUREBERG"
    if "GILLBO" in u_region:
        return "GILLBO"
    if "EDSÄNGEN" in u_region:
        return "EDSÄNGEN"
    if "LANDSNORA" in u_region:
        return "LANDSNORA"
    if "VAXMORA" in u_region:
        return "VAXMORA"
    if "SOLÄNGEN" in u_region:
        return "SOLÄNGEN"
    if "VÄSJÖN" in u_region:
        return "VÄSJÖN"
    if "KVARNSKOGEN" in u_region:
        return "KVARNSKOGEN"
    return "SOLLENTUNA"

df['region'] = df.apply(lambda row: clean_region(row['region']), axis=1)
df.region.unique()
```

```python
df.region.value_counts()
```

```python
df.broker.fillna('UNKNOWN', inplace=True)

def clean_broker(broker):
    if "Fastighetsbyrån" in broker:
        return "Fastighetsbyrån"
    if "Mäklarhuset" in broker:
        return "Mäklarhuset"
    if "Bjurfors" in broker:
        return "Bjurfors"
    if "Susanne Persson" in broker:
        return "Susanne Persson"
    if "Notar" in broker:
        return "Notar"
    if "Länsförsäkringar" in broker:
        return "Länsförsäkringar"
    if "HusmanHagberg" in broker:
        return "HusmanHagberg"
    if "Svensk Fastighetsförmedling" in broker:
        return "Svensk Fastighetsförmedling"
    if "SkandiaMäklarna" in broker:
        return "SkandiaMäklarna"
    if "ERA" in broker:
        return "ERA"
    if "Erik Olsson" in broker:
        return "Erik Olsson"
    return "Other"

df['broker'] = df.apply(lambda row: clean_broker(row['broker']), axis=1)
df.broker.value_counts()
```
