# Understanding the Missing Values

## Address

```python
df[pd.isnull(df['address'])]
```

```python
df[pd.isnull(df['region'])]
```

This is crucial for training so we will have to drop the ones with missing area

```python
df[pd.isnull(df['area'])]
```
